```python
# Imports
import numpy as np
import pandas as pd
import plotly.express as px

# Inline Plots
%matplotlib inline
```

```python
# Read-in .csv file
sunspots_dataset = pd.read_csv('./data/sunspots.csv')

# Get dataframe info
sunspots_dataset.info()
```

```python
# Drop unnecessary column:
sunspots_dataset = sunspots_dataset.drop(columns=['Unnamed: 0'])
```

```python
# Preview dataframe
sunspots_dataset.head()
```

```python
# Rename columns
sunspots_dataset.columns = ['Year', 'Monthly Mean']
```

```python
plot = px.line(sunspots_dataset, x=sunspots_dataset.index, y='Monthly Mean', title='Sunspots Monthly Mean')
plot.show()
```
